## üî•‚ùÑÔ∏è Efeito Mpemba 
O curioso fen√¥meno no qual √°gua quente congela mais r√°pido √© conhecido como efeito Mpemba, nome do estudante tanzaniano que o observou ao preparar sorvete e o lan√ßou como desafio para a comunidade cient√≠fica na d√©cada de 60. Historicamente,ind√≠cios de que a √°gua quente congela mais r√°pido que a √°gua fria remontam √† √©poca de Arist√≥teles, com observa√ß√µes posteriores feitas por Bacon, Joseph Black e at√© Fahrenheit. Erasto Mpemba observou o efeito em uma aula de culin√°ria e, intrigado com a f√≠sica envolvida no fen√¥meno, questionou destemidamente um palestrante que visitou sua escola. Desde o artigo ["Cool?"](https://iopscience.iop.org/article/10.1088/0031-9120/4/3/312), co-autorado por Mpemba e o professor Denis Osborne, o fen√¥meno passou a ser conhecido como o efeito Mpemba. As aplica√ß√µes v√£o al√©m dos desafios de sobremesas congeladas no programa MasterChef. 

Cientificamente, o efeito Mpemba ainda √© alvo de debate, especialmente porque sua manifesta√ß√£o em escalas macrosc√≥picas n√£o possui uma explica√ß√£o termodin√¢mica satisfat√≥ria. No entanto, em escalas microsc√≥picas, o efeito parece ser muito mais evidente, o que tem gerado um crescente interesse em compreender o chamado efeito Mpemba qu√¢ntico. N√£o demorou para que a comunidade de termodin√¢mica qu√¢ntica se debru√ßasse sobre este desafio, buscando identificar os ingredientes e uma "receita" para ativ√°-lo em sistemas qu√¢nticos, tais como qubits. Curiosamente, essa mesma receita pode ajudar a explicar o cl√°ssico caso da √°gua quente congelando mais r√°pido que a fria.


### ‚è≥ Mpemba cl√°ssico versus qu√¢ntico üßò‚Äç‚ôÄÔ∏è
No contexto de sistemas cl√°ssicos ou qu√¢nticos, pode-se atribuir ao efeito Mpemba qualquer situa√ß√£o de relaxa√ß√£o acelerada üßò‚Äç‚ôÄÔ∏è. No caso do sorvete, esse processo de relaxa√ß√£o nada mais √© do que a termaliza√ß√£o com o freezer. Imediatamente quando a massa de sorvete entra no freezer, seja quente ou a temperatura ambiente $T_0 = 25^oC$, ela est√° aquela temperatura. Podemos abrir o freezer diversas vezes em intervalos de tempo $\Delta t$ de 10 minutos e medir a temperatura da massa üå°Ô∏è at√© ele congelar completamente (n√£o recomendo fazer isso porque a abertura do freezer vai fazer com que o sorvete demore ainda mais pra congelar kkkk). Em cada um desses instantes, se f√≥ssemos capazes de medir a temperatura da massa v√°rias vezes, ver√≠amos que ela est√° ficando cada vez menor $T((n+1)\Delta t) < T(n\Delta t)$, onde $n$ √© a vez em que o freezer foi aberto (primeira vez, segunda, ..., √∫ltima). No final, em $t = N \Delta t$ quando o sorvete congelou, ele est√° exatamente na mesma temperatura do freezer $T(N\Delta t) = T_f$, pode ser, por exemplo, $-8^oC$. Vamos dizer que demorou $t = 2h$ para que ele congelasse completamente. No Mpemba sorveteiro, se coloc√°ssemos a mesma massa a $T_0'= 50^oC$, observar√≠amos que o tempo total para que o sorvete ficasse pronto seria menor! Isto √©, seria menor do que as $2h$ do caso onde ele entrou no freezer quente. Por exemplo, poderia ter tomado $t' = 1h20$.  


Essa mesma ideia vale para o aquecimento: pode acontecer que um preparo inicial a temperatura ambiente demore mais tempo para esquentar do que um preparo congelado.

Veja que esse processo de sair de um estado inicial (sorvete a temperatura ambiente ou quente) e chegar a um estado final (sorvete congelado) √© um processo de relaxa√ß√£o üßò‚Äç‚ôÄÔ∏è. A temperatura üå°Ô∏è √© a quantidade que nos indica quando o sistema alcan√ßou o estado final desejado. O tempo ‚è≥ reduzido de um dos estados iniciais que tinha aquele par√¢metro maior no come√ßo (no caso, a temperatura üå°Ô∏è) √© um indicativo do Mpemba. Neste caso, podemos tomar a temperatura como um bom quantificador de dist√¢ncia entre o estado inicial e o final. O estado mais quente inicialmente tinha uma dist√¢ncia maior do estado final desejado, enquanto o a temperatura ambiente tinha uma dist√¢ncia menor. Se um estado inicial com dist√¢ncia maior relaxa mais r√°pido, voi-l√°, aconteceu o Mpemba!

Podemos extrapolar esse conceito de relaxa√ß√£o acelerada para diversas situa√ß√µes. üë©‚Äçüî¨ Cientificamente, podemos pensar em acelerar uma rea√ß√£o qu√≠mica, um processo de polimeriza√ß√£o, etc, etc. No caso qu√¢ntico ‚öõÔ∏è, podemos buscar acelerar rotas para fases ex√≥ticas da mat√©ria, ou mesmo para resfriar sistemas qubits em um computador qu√¢ntico. 




## üë©‚Äçüç≥ Receita para o Mpemba Qu√¢ntico 

No artigo ["Thermodynamics of the quantum Mpemba effect"](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.133.140404), eu, em parceria com Mattia Moroder, Oisin Culhane e John Goold do Trinity College Dublin, propomos uma receita para fazer "sorvete de qubits". A est√≥ria de como esse artigo surgiu durante meu posdoc no grupo √© hil√°ria e merece uma prosa com algumas bolas de sorvete ...


Sucintamente, nos intrigou a ideia de que deveria existir alguma propriedade termodin√¢mica que seria a chave para entender como o efeito se manifesta. Sendo capaz de manipular esta propriedade convenientemente, gostar√≠amos de ativar o efeito independente das condi√ß√µes iniciais do sistema. Veja que, a emerg√™ncia do Mpemba s√≥ acontece para classes especiais de estados iniciais. Ent√£o, pensamos que bastaria fazer algo com o estado inicial para que ele se tornasse o mais pr√≥ximo poss√≠vel dos estados que manifestariam o Mpemba. 

### Quantidades termodin√¢micas
No exemplo do sorvete, utilizamos a temperatura como um quantificador de dist√¢ncia. H√° um problema com a temperatura durante o congelamento ... h√° partes da massa de sorvete que ainda podem estar l√≠quidas, enquanto outras est√£o j√° formando cristais de gelo. Nessa situa√ß√£o, dizemos que o sistema est√° fora do equil√≠brio üòµ‚Äçüí´, ele n√£o relaxou ainda ... Al√©m disso, n√£o h√° uma √∫nica temperatura para caracterizar a massa toda e, talvez, a m√©dia n√£o seja um bom quantificador. H√° uma outra quantidade termodin√¢mica que nos permite melhor caracterizar o estado instant√¢neo do sistema para quantificar a dist√¢ncia: a **energia**! Mais precisamente, a **energia livre fora de equil√≠brio**, que denotamos por $F_{\textrm{neq}}$. Ela tamb√©m permite ''medir'' dist√¢ncias entre estados, seja no caso cl√°ssico ou qu√¢ntico.


### Termaliza√ß√£o
Ainda no exemplo do sorvete, usamos um freezer como um agente externo capaz de retirar energia t√©rmica da massa para que ele congele. Poder√≠amos ter esquentado sopa tamb√©m ... Tanto o freezer quanto o fogo s√£o o que chamamos de banho ou reservat√≥rio t√©rmico. Para modelar um processo de resfriamento/aquecimento, precisamos entender como esse banho interage com o sistema e modifica o estado interno dele. Essa descri√ß√£o no contexto de f√≠sica qu√¢ntica √© dada pela **teoria de sistemas qu√¢nticos abertos** ‚úçÔ∏è. 

Uma das constru√ß√µes dessa teoria √© uma equa√ß√£o mestra para governar a evolu√ß√£o de um sistema interagindo com um reservat√≥rio ou v√°rios reservat√≥rios, sejam eles t√©rmicos ou n√£o. Por exemplo, essa teoria nos ajuda a entender como o ru√≠do do ambiente corrompe as computa√ß√µes de um algoritmo rodando em um computador qu√¢ntico. Nossos computadores qu√¢nticos atuais precisam ficar refrigerados a temperaturas baix√≠ssimas, al√©m de serem super dooper isolados do ambiente, para que as opera√ß√µes n√£o sejam corrompidas. Por isso, dizemos que estamos na era NISQ (noisy intermediate scale quantum devices), a era dos computadores qu√¢nticos ruidosos. Neste exemplo, o reservat√≥rio √© bem complexo ...

Voltando a como a teoria modela a intera√ß√£o entre o reservat√≥rio e o sistema: ela nos diz que podemos representar efetivamente o efeito do banho com um **super operador** que nada mais faz do que promover transi√ß√µes entre os estados internos do sistema a taxas que dependem da diferen√ßa de energia entre estes estados internos. Os estados internos que o sistema pode ter s√£o codificados no seu **Hamiltoniano**, denotado por $\hat{H}$, um operador qu√¢ntico associado √† energia do sistema. 

Esse super operador, denominado **Liouviliano**, cont√©m esses **operadores de jump**, denotados por $\hat{L}_j$ ($j$ pode indicar v√°rios operadores), que promovem essas transi√ß√µes. 

Assim, estaremos interessados em resolver a seguinte din√¢mica:
$$ \frac{d\hat{\rho}(t)}{dt}=\mathcal{G}\hat{\rho}(t), $$
onde o super operador Liouvilliano cont√©m uma contribui√ß√£o da din√¢mica do pr√≥prio sistema dada por $\hat{H}$ e outra que depende do banho
$$
\mathcal{G}[.] = \underbrace{-i [\hat{H}, . ]}_{\textrm{unitaria}} + \underbrace{\mathcal{\, \, \, D}(.)\, \,\, }_{\textrm{reservat√≥rio}},
$$
e
$$
\displaystyle  \mathcal{D}(\mathord{\cdot}) =  \sum_\omega  \gamma(\omega) \left[
		\hat{L}(\omega) (.) \hat{L}^\dagger(\omega) - 
		 \frac{1}{2} \{ \hat{L}^\dagger (\omega) \hat{L} (\omega),  . \} \right],
$$
onde $\gamma(\omega)$ representa a taxa de transi√ß√£o.

A determina√ß√£o destes operadores e taxas para o caso espec√≠fico de um banho t√©rmico √© chamado de **mapa de Davies**. Vamos usar o mapa de Davies para fazer o sorvete de um qubit.


### Manipula√ß√£o do estado inicial
Matematicamente, alterar o estado de um qubit √© an√°logo a aplicar uma transforma√ß√£o em uma imagem para gerar duas cenas diferentes de uma anima√ß√£o. Um frame da anima√ß√£o seria a entrada inicial e uma sequ√™ncia de frames a partir dele √© gerada atrav√©s de multiplica√ß√µes matriciais. Por exemplo, para girar um objeto em uma cena, basta aplicar rota√ß√µes no objeto. √â exatamente isto que fazermos com sistemas qu√¢nticos: uma **transforma√ß√£o** $\hat{U}$ que leva um estado inicial $\hat{\rho}(t)$ em um outro estado $\hat{\rho}'$ nada mais √© do que uma **rota√ß√£o**, implementada via 

$$\hat{\rho}' = \hat{U} \hat{\rho} \hat{U}^\dagger.$$  

N√≥s identificamos que h√° uma rota√ß√£o especial que pode gerar o Mpemba para qualquer estado inicial: ela √© decomposta em duas rota√ß√µes: $\hat{U} = \hat{U}_2 \hat{U}_1$. A primeira, $\hat{U}_1$ elimina as coer√™ncias de um estado inicial. A segunda, $\hat{U}_2$ ordena as popula√ß√µes deixando o estado final mais pr√≥ximo do desejado ap√≥s a relaxa√ß√£o. 

Mas, mas, mas, o que s√£o coer√™ncias e popula√ß√µes?? D√° um exemplo de Hamiltoniano e banho, please! Vamos explicar tudo isto no exemplo a seguir.


## Exemplo: um qubit
Bem, vamos tomar o exemplo de um qubit: um qubit √© sistema que pode estar em uma combina√ß√£o de dois estados $|0 \rangle$ e $|1 \rangle$. Pode haver uma probabilidade $p_{00}$ de que o qubit esteja no estado $|0 \rangle$ e $p_{11}$ de estar em $|1 \rangle$. Pode, ainda haver transi√ß√µes entre cada um destes estados. Uma transi√ß√£o de de $|1 \rangle$ para $|0 \rangle$ √© codificada como $|0 \rangle \langle 1|$. Pode-se atribuir uma probabilidade $p_{10}$ para que ela aconte√ßa tamb√©m. Assim, o estado de um qubit √© uma matriz hermitiana $2\times 2$, escrita como

$$
\hat{\rho} = \begin{pmatrix}
p_{00} & p_{01}\\
p_{10} & p_{11}
\end{pmatrix}.
$$

As entradas diagonais $p_{00}$ e $p_{11}$ s√£o as popula√ß√µes, enquanto as n√£o-diagonais $p_{01}$ e $p_{10}$ s√£o chamadas coer√™ncias.

O estado de um qubit tamb√©m pode ser codificado em um vetor tridimensonal $\vec{r} = (r_x, r_y, r_z)$ que roda a chamada esfera de Bloch. Nessa representa√ß√£o tridimensional, o p√≥lo norte da esfera equivale ao estado $\ket{0}$, o sul ao estado $\ket{1}$, enquanto que qualquer estado com longitude e latitude √© uma mistura como a descrita acima. 


O Hamiltoniano tamb√©m ser√° uma matriz hermitiana $2 \times 2$, 
$$
\hat{H} = \begin{pmatrix}
h_{00} & h_{01}\\
h_{10} & h_{11}
\end{pmatrix}.
$$

Ao diagonalizar o Hamiltoniano, obteremos os estados poss√≠veis $|\epsilon_0 \rangle$e $|\epsilon_1 \rangle$ para o qubit, com energias $\epsilon_0$ e $\epsilon_1$, respectivamente.

Os jump operators promover√£o transi√ß√µes entre estes n√≠veis e, portanto, tamb√©m ser√£o matrizes $2 \times 2$
$$
\hat{L}^\pm = \begin{pmatrix}
l_{00}^\pm & l_{01}^\pm\\
l_{10}^\pm & l_{11}^\pm
\end{pmatrix},
$$
com taxas $\gamma^+(\omega = \epsilon_1 - \epsilon_0)$ e $\gamma^+(\omega = \epsilon_0 - \epsilon_1)$.

Pronto! Temos tudo o que precisamos para modelar a evolu√ß√£o. Por√©m, resolver a equa√ß√£o mestra numericamente pode ser muito mais convenientemente implementado se **vetorizarmos o gerador**. Um super operador, como o pr√≥prio nome diz, n√£o √© um operador e, assim, n√£o √© uma simples matriz. Mas, ele pode se tornar uma! Se vetorizarmos a matriz de densidade
$$
 \hat{\rho} = \sum_{ij} \rho_{ij} \ket{i} \bra{j} 
		\rightarrow \ket{\rho}\!\rangle = \sum_{ij} \rho_{ij} \ket{i} \ket{j},
$$
podemos re-escrev√™-lo como uma matriz $4 \times 4$.

$$
 \displaystyle
       \hat{\mathcal{G}} = -i \hat{H} \otimes  \hat{\mathcal{1}} + i   \hat{\mathcal{1}} \otimes \hat{H}^{\scriptscriptstyle T}
 \hphantom{G} +\sum_l \hat{L}_l \otimes \left ( \hat{L}^{\dagger}_l \right ) ^{\scriptscriptstyle T} \! \! -\frac{1}{2} \left[ \hat{L}^{\dagger}_l \hat{L}_l \otimes  \hat{\mathcal{1}}  + \hat{\mathcal{1}} \otimes \!\left( \hat{L}^{\dagger}_l \hat{L}_l \right) ^{\scriptscriptstyle T}\right].
$$

Maravilha, agora, se diagonalizarmos esta matrix $\mathcal{G}$, que n√£o √© mais Hermitiana, obteremos auto valores $\lambda_k$ complexos e, p√°smem, auto-vetores √† esquerda 
$$
\mathcal{G}^{\dagger}[\hat{l}_{k}]=\lambda_{k}\hat{l}_{k}
$$
e √† direita
$$
\mathcal{G}[\hat{r}_{k}]=\lambda_{k}\hat{r}_{k}.
$$

A vantagem √© que, agora, a evolu√ß√£o antes modelada pela equa√ß√£o mestra √© descrita em termos destes auto vetores e autovalores
$$
\hat{\rho}(t) = e^{\mathcal G t} \hat{\rho}(0) = \sum_{k=1}^{K}\Tr(\hat{l}_{k}\hat{\rho}_i)\hat{r}_ke^{\lambda_k t},
$$
onde $\hat{\rho}(0)$ √© o estado inicial.

````python
# tempo de simula√ß√£o da termaliza√ß√£o
t_max = 8
dt = 0.01
time_v = np.arange(0, t_max+dt, dt)
Nt = len(time_v)

# estados instant√¢neos
state_original_vectorized_every_t = np.zeros((4**n_sites,Nt), dtype=np.complex128)
state_diagonalized_vectorized_every_t = np.zeros((4**n_sites,Nt), dtype=np.complex128)

# estado estacion√°rio
id_v = np.eye(2**n_sites).flatten()
steady_state = np.reshape(right_eigv_vectorized_lindbladian[:,0], (2**n_sites,2**n_sites))
steady_state /= np.trace(steady_state)

# quantidades termodin√¢micas
F_original = np.zeros((Nt))
F_rotated = np.zeros((Nt))

distance_original = np.zeros((Nt))
distance_rotated = np.zeros((Nt))

# din√¢mica
for i, t in enumerate(time_v):
    # din√¢mica dos dois estados
    state_original_vectorized = scipy.linalg.expm(t*lindbladian_vectorized) @ original_state_original_vectorized
    state_diagonalized_vectorized = scipy.linalg.expm(t*lindbladian_vectorized) @ original_state_diagonalized_vectorized
    state_original = np.reshape(state_original_vectorized, (2**n_sites,2**n_sites))
    state_diagonalized = np.reshape(state_diagonalized_vectorized, (2**n_sites,2**n_sites))
    state_original /= np.trace(state_original)
    state_diagonalized /= np.trace(state_diagonalized)
    
    state_original_vectorized_every_t[:,i] = state_original_vectorized
    state_diagonalized_vectorized_every_t[:,i] = state_diagonalized_vectorized
    
    # energia livre de cada estado
    F_original[i] =  noneq_free_energy(state_original, np.diag(eigw_hamiltonian), temperature_bath) 
    F_rotated[i] = noneq_free_energy(state_diagonalized, np.diag(eigw_hamiltonian), temperature_bath) 
    
    # dist√¢ncias entre estados - L1 distance 
    distance_original[i] =  np.sum(np.abs(state_original - steady_state)) 
    distance_rotated[i] =  np.sum(np.abs(state_diagonalized - steady_state))    
    
````
